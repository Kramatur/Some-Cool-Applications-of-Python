```python
  ###############################################Download Single YouTube Video################################################

#import necessary library (Modules) 
#!pip install pytube --if you haven't installed the 'pytube' library, intall the library first
#!pip install emoji  --if you haven't installed the 'emoji' library, intall the library first

from pytube import YouTube
import emoji 

# Set the path where you want to save the video  
# (Specifying the location for this video)
Path = "C:/Users/kramatur.r/Downloads/Video" #Specify your desired path

#Ask user to enter YouTube video link wants to download
link = input("Enter the link of youtube video want to download:  ")

#Creating an object for download

try:  
    # object creation using YouTube 
    # which was imported in the beginning  
    yt = YouTube(link)  
except:  
    print("Connection Error") # Exception Handling
    
#To get the highest resolution
Video = yt.streams.get_highest_resolution()

#Show the message until downloading
print("Downloading is being processing...")

try:  
    # downloading the video  
    Video.download(Path)  
except:  
    print("Some Error!")  
    #Show the message when download is completed
print(emoji.emojize("Congrates, Your video is ready to watch :grinning_face_with_big_eyes:"))  
```

```python
 ###############################################Download Multiple YouTube Video###############################################
    

#import necessary library (Modules) 
#pip install pytube --if you haven't installed the 'pytube' library, intall the library first
#pip install emoji  --if you haven't installed the 'emoji' library, intall the library first

from pytube import YouTube
import emoji 

# Set the path where you want to save the video  
# (Specifying the location for this video)
Path = "C:/Users/kramatur.r/Downloads/Video" #Specify your desired path

#Ask user to enter YouTube video link wants to download
link = []
links = int(input("Enter the Number of video links you want to download:  "))
print("Enter the links:  ")
for i in range(links):
    videolinks = input()
    link.append(videolinks)
print(link)
#creating an object for download
for j in link:
    try:  
        # object creation using YouTube  
        yt = YouTube(j)  
    except:  
        print("Connection Error") # Exception Handling   

    #To get the highest resolution
    Video = yt.streams.get_highest_resolution()

    #Show the message until downloading
    print("Downloading is being processing...")

    try:  
        # downloading the video  
        Video.download(Path)  
    except:  
        print("Some Error!") 

        #Show the message when download is completed
    print(emoji.emojize("Congrates, Your video is ready to watch :grinning_face_with_big_eyes:"))  
```
